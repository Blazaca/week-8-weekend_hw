```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?
n_groups = air_bnb.groupby(by=['neighbourhood_group'], as_index=True).count()
n_groups.sort_values(by=['id'], ascending=False)
```

```python
# Are private rooms the most popular in manhattan?
manhat_priv = air_bnb[['neighbourhood_group', 'room_type']]
manhat_priv = manhat_priv.groupby(['neighbourhood_group', 'room_type']).size().sort_values(ascending=False)
print(manhat_priv.filter(like='Private room', axis=0))
print('\nBrooklyn has the most private rooms in the area.')
```

```python
# Which hosts are the busiest and based on their reviews?
max_reviews = air_bnb.groupby(['host_id']).count()
max_reviews = max_reviews.sort_values(by=['number_of_reviews'], ascending=False)
print(max_reviews['number_of_reviews'].head(5))
```

```python
#Which neighorhood group has the highest average price?
avg_price = air_bnb[['neighbourhood_group', 'price']]
avg_price = avg_price.groupby(['neighbourhood_group']).mean()
# avg_price.sort_values(by=['mean'], ascending=False)
print(avg_price.sort_values(by=['price'], ascending=False).head(1))
```

```python
# Which neighbor hood group has the highest total price?
total_price = air_bnb[['neighbourhood_group', 'price']]
print(total_price.groupby(['neighbourhood_group']).sum().sort_values(by=['price'], ascending=False).head(1))
```

```python
#Which top 5 hosts have the highest total price?
top_prices = air_bnb[['host_id', 'price']]
print(top_prices.groupby(['host_id']).sum().sort_values(by=['price'], ascending=False).head(5))
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
avail_bnb = air_bnb[['host_id', 'number_of_reviews', 'availability_365']]
avail_bnb = avail_bnb.loc[(avail_bnb['availability_365']==0)&(avail_bnb['number_of_reviews']>=100)]
print(avail_bnb.sort_values(by=['number_of_reviews'], ascending=False))
print('\n All 162 of these hosts fit the description')
```

```python
# What host has the highest total of prices and where are they located?
total_price_location = air_bnb[['host_id', 'price', 'neighbourhood_group']]
total_price_location.groupby(['host_id'], as_index=True).sum().sort_values(by=['price'], ascending=False)
air_bnb.loc[(air_bnb['host_id']==219517861)].groupby(['neighbourhood_group']).count()
print('Host: 219517861 holds the max for highest total cost in the Manhattan area')
```

```python
# When did Danielle from Queens last receive a review?
dan_queen = air_bnb.loc[(air_bnb['host_name']=='Danielle') & (air_bnb['neighbourhood_group']=='Queens')]
dan_queen = dan_queen.sort_values(by=['last_review'], ascending=False)
dan_queen[['host_name', 'last_review']].head(1)
```

## Further Questions

1. Which host has the most listings?

```python
max_list = air_bnb[['host_id', 'id']].groupby(['host_id']).count().sort_values(by=['id'], ascending=False).head(1)
print(max_list)
```

2. How many listings have completely open availability?

```python
open_aval = air_bnb.loc[(air_bnb['availability_365']==365)]
print(open_aval[['id']].count())
print('\nThere are 1295 listings with open availability')
```

3. What room_types have the highest review numbers?

```python
high_numbers = air_bnb[['room_type', 'number_of_reviews']].groupby(['room_type']).count()
print(high_numbers)
print('\n Entire homes and apartments have the highest total')
```
